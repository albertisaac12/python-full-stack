```python
def longestCommonPrefix(strs: list) -> str:
    if len(strs) == 0:
        return ""

    if len(strs) == 1:
        return strs[0]

    strs.sort()

    l1 = len(strs[0])
    l2 = len(strs[1])

    count = 0
    for i in range(min(l1, l2)):
        if strs[0][i] == strs[1][i]:
            count += 1
        else:
            break

    return strs[0][:count]
```

```python
print(longestCommonPrefix(["ab", "a"]))
```
